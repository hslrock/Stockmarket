### Numpy Random Walk Generation

```python
import numpy as np

price = 1

price_list = []
for _ in range(1000):
    price_list.append(price)
    price += np.random.randn()
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8))
plt.plot(price_list)
plt.grid(True)
plt.show()
```

```python
x_full=df['Close'].values
```

```python
num_steps=15
scale_factor=0.7
val=len(x_full)
size_list=[]
index_list=[]
for i in range(num_steps):
    size_list.append(val)
    val=int(val*(scale_factor))
x_pyramids=[]
for size in size_list:
    
    index=list(np.linspace(start=0, stop=len(x_full)-1, num=size,dtype=int))
    index_list.append(index)
    x_pyramids.append(x_full[index])
plt.figure(figsize=(20,10))
for i in range(len(index_list)):
    plt.plot(index_list[i],x_pyramids[i])
x_pyramids.reverse()
```

```python
#Training Process

scale_num=0
max_stop_scale=6
Gs=[]
Zs=[]
NoiseAmp = []
scale=0
in_s = 0
reals=x_pyramids
D_steps=3
G_steps=3
alpha=10
```

```python
from networks.model import init_models
from networks.singan_train import train_single_scale
from utils import *
```

```python
device='cuda'
while scale_num<len(reals):
    #print(scale_num)
    D_curr,G_curr=init_models()
    D_curr.to(device)
    G_curr.to(device)
    z_curr,in_s,G_curr,noise_amp = train_single_scale(D_curr,G_curr,reals,Gs,Zs,in_s,NoiseAmp,'cuda',D_steps,G_steps,alpha)
    G_curr = reset_grads(G_curr,False)
    G_curr.eval()
    D_curr = reset_grads(D_curr,False)
    D_curr.eval()
    Gs.append(G_curr)
    Zs.append(z_curr)
  #  print(z_curr)
    NoiseAmp.append(noise_amp)
    scale_num+=1
    del D_curr,G_curr
```
