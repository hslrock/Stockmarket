```python
import numpy as np

price = 1

price_list = []
for _ in range(1000):
    price_list.append(price)
    price += np.random.randn()

import matplotlib.pyplot as plt

plt.figure(figsize=(16,8))
plt.plot(price_list)
plt.grid(True)
plt.show()
```
